```python
jr_sorted  = joined_results.sort('BICCN_cluster_label')

list1 = jr_sorted['BICCN_cluster_label']
list2 = jr_sorted['class_name']
label_counts = {}
mapping_percentages = {}

# Iterate through the lists and calculate label counts
for item1, item2 in zip(list1, list2):
    if item1 in label_counts:
        label_counts[item1] += 1
    else:
        label_counts[item1] = 1

# Iterate through the lists again and calculate mapping percentages
for item1, item2 in zip(list1, list2):
    if (item1, item2) not in mapping_percentages:
        mapping_percentages[(item1, item2)] = (1 / label_counts[item1]) * 100

# Display the mapping percentages
for mapping, percentage in mapping_percentages.items():
    print(f"Mapping {mapping[0]} to {mapping[1]} = {percentage:.2f}%")
```
